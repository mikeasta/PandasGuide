У какой доли студентов из датасэта в колонке lunch указано free/reduced?

```python
overall_amount = len(students_performance)
free_amount = len(students_performance.loc[students_performance.lunch == "free/reduced"])
ratio = free_amount / overall_amount
ratio
```

Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?

```python
free_mean = students_performance.loc[students_performance.lunch == "free/reduced"]['writing score'].mean()
free_diff = students_performance.loc[students_performance.lunch == "free/reduced"]['writing score'].var()

st_mean = students_performance.loc[students_performance.lunch == "standard"]['writing score'].mean()
st_diff = students_performance.loc[students_performance.lunch == "standard"]['writing score'].var()

print(f'Free/Reduced: {free_mean}, {free_diff}. Standard: {st_mean}, {st_diff}')
```
